```python
class operation:
    def __init__(self):
        self.myclient = pymongo.MongoClient("mongodb://localhost:27017/")
        self.mydb = self.myclient["zomato"]
        self.col = self.mydb['hotels']
        
    def conn(self):
        print("calling connection")
        #myclient = pymongo.MongoClient("mongodb://localhost:27017/")
        #mydb = myclient["zomato"]
        #col = mydb['hotels']
        x = self.col.find()

        self.myclient.close()
        
    def first_query(self, get, rating):
        #myclient = pymongo.MongoClient("mongodb://localhost:27017/")
        #mydb = myclient["zomato"]
        #col = mydb['hotels']
        restaurants = self.col.find({"Aggregate rating":{"$gte":float(rating)},  "City": {'$regex': get, '$options':'i'}})
        results_rating_for_city = []
        for i in restaurants:
            results_rating_for_city.append(i)
        first_dataframe = pd.DataFrame(results_rating_for_city)
        first_dataframe = first_dataframe[['Restaurant Name','Address','Votes','Average Cost for two','Aggregate rating']]
        first_dataframe = first_dataframe.sort_index(ascending=False)
        #print(re)
        return first_dataframe
    
    def second_query(self, get):
        #myclient = pymongo.MongoClient("mongodb://localhost:27017/")
        #mydb = myclient["zomato"]
        #col = mydb['hotels']
        second_dataframe = pd.DataFrame(list(self.col.find({'Cuisines':{'$regex': get, '$options':'i'}})))
        second_dataframe = second_dataframe[['Restaurant Name','Cuisines','Address','City','Average Cost for two','Aggregate rating']]
        second_dataframe = second_dataframe.sort_index(ascending=False)
        
        #list_element = pd.DataFrame(list(self.col.find()))
        #list_element = list_element.sort_index(ascending=False)
        #print(list_element)
        
        return second_dataframe
    
    def third_query(self):
        '''In the datframe cuisines for the restaurants are given in single column we are having a task
        to separate them and show only unique cuisines over the whole dataset
        '''
        cuisine_list = self.col.distinct("Cuisines")
        cuisines = [i.split(',')[0] for i in cuisine_list]
        results = []
        for i in cuisines:
            if i not in results:
                results.append(i)
                
        cuisines_data = pd.DataFrame(results, columns=['Cuisines'])
        cuisines_data = cuisines_data.sort_index(ascending= False)
        return cuisines_data
    
    def fourth_query(self, get):
        '''User will provide city name for which he want's the average cost for two '''
        get_dataframe = self.third_query()
        get_dataframe = get_dataframe['Cuisines'].values.tolist()
        get_indexes = []
        for cuisine in get_dataframe:
            get_indexes.append(self.col.aggregate([{'$match':{'$and':[{'Cuisines':{'$regex':cuisine,'$options':'i'}}, 
                                                                      {'City':{'$regex':get,'$options':'i'}}]}},{'$group': {'_id':{'City':'$City','Cuisine':cuisine}, 'Average cost':{'$avg':'$Average Cost for two'}}}, {'$sort':{'Average cost':1}}]))
            
        result_dataframe = pd.DataFrame(list(get_indexes), columns=['Cuisines'])
        result_dataframe = result_dataframe.dropna()
        result_dataframe = result_dataframe.reset_index(drop=True)
        result_dataframe = result_dataframe['Cuisines'].apply(pd.Series)
        collect_avg_cost = result_dataframe
        result_dataframe = result_dataframe['_id'].apply(pd.Series)
        for_print = pd.concat([result_dataframe['Cuisine'],collect_avg_cost['Average cost']], axis = 1)
        for_print = for_print.sort_values(by=['Average cost'])
        for_print = for_print.reset_index(drop= True)
        for_print = for_print.sort_index(ascending= False)
        
        return for_print
    
    def fifth_query(self):
        cities_list = pd.DataFrame(list(self.col.aggregate([{'$group':{'_id':'$City'}},{'$sort':{'City':1}}])))
        cities_list.rename(columns={'_id':'CITY'}, inplace=True)
        
        cities_list = cities_list.sort_values('CITY', ascending = True).reset_index()
        cities_list = cities_list.sort_index(ascending= False)
        cities_list = cities_list[['CITY']]
        cities_list_to_check = cities_list.values.tolist()
        print(cities_list_to_check)
        flatList = [ item for elem in cities_list_to_check for item in elem]
        count_list = []
        results = self.third_query()
        results = results['Cuisines'].values.tolist()
        
        for city in flatList:
            for cuisine in results:
                count_list.append(self.col.aggregate([{'$match':{'$and':[{'Cuisines':{'$regex':cuisine, '$options':'i'}}, {'City':{'$regex':city,'$options':'i'}}]}},
                                        {'$group':{'_id':{'Cuisine':cuisine,'City':'$City'}, 'Vote Count':{'$sum':'$Votes'} }}]))
        
        
        
        
        df = pd.DataFrame(list(count_list), columns=['List'])
        df2 = df['List'].apply(pd.Series)
        df2 = df.dropna()
        df3 = df2['List'].apply(pd.Series)
        df4 = df3['_id'].apply(pd.Series)
        print(df4)
        df6 = df3['Vote Count']
        df5 = pd.concat([df4,df6], axis = 1)
        check = df5.loc[df5.groupby('City')['Vote Count'].idxmax()]
        check = check[['City','Cuisine','Vote Count']]
        
        check = check.reset_index()
        del check['index']
        
        fifth_query_dataframe  = check.sort_index(ascending= False)
        
        print('fifth_query_completed')
        
        
        
        
        
        
        
        
        '''In this will take only rating from user and kept voting count greater than 100 default'''
        '''fifth_query_list =  self.col.find({'$and':[{'City':{'$regex':city_val, '$options':'i'}},{'Aggregate rating':{'$gte':float(rating_val)}},
                                                        {'Votes':{'$gte':100}}]}).sort([('Aggregate rating',-1),('Average Cost for two', 1)])
        fifth_query_dataframe = pd.DataFrame(list(fifth_query_list))

        fifth_query_dataframe = fifth_query_dataframe[['Restaurant Name','Average Cost for two','Aggregate rating']]
        fifth_query_dataframe = fifth_query_dataframe.sort_index(ascending = False)'''
        
        return fifth_query_dataframe
    
    def sixth_query(self, get):
        '''Finding restaurants all over the world from dataset and will take the count for minimum restaurants outlet
           from user
        '''
        sixth_query_list = self.col.aggregate([{'$group':{'_id':'$Restaurant Name', 'count':{'$sum':1}, 'rating':{'$avg':{'$ifNull':['$Aggregate rating',0]}}}},{'$match':{'count':{'$gt':int(get)}}},{'$sort':{'count':-1}}])
        sixth_query_dataframe = pd.DataFrame(list(sixth_query_list))
        sixth_query_dataframe.rename(columns = {'_id':'Restaurant Name','count':'Number of Outlets', 'rating':'Avaerage Rating'}, inplace= True)
        
        sixth_query_dataframe = sixth_query_dataframe.sort_index(ascending = False)
        
        return sixth_query_dataframe
    
    def seventh_query(self, city_val, cuisine_val):
        '''Checking option for table booking with rating and cost provided'''
        seventh_query_list = self.col.find({'$and':[{'Cuisines':{'$regex': cuisine_val, '$options':'i'}}, {'City':{'$regex':city_val, '$options':'i'}},
                                      {'Has Table booking':'Yes'}]}).sort([('Aggregate rating',-1),('Average Cost for two',1)])
        seventh_query_dataframe = pd.DataFrame(list(seventh_query_list))
        
        seventh_query_dataframe = seventh_query_dataframe[['Restaurant Name','Address','Has Table booking','Average Cost for two','Aggregate rating']]
        
        seventh_query_dataframe = seventh_query_dataframe.sort_index(ascending= False)
        
        return seventh_query_dataframe
    
    def eight_query(self, city_val, cuisine_val, price_val):
        '''Getting unique list of cities witout distinct keyword as we know 'distinct' keyword takes high i/o access '''

        
        match_hotels = self.col.aggregate([{'$match':{'$and':[{'Cuisines':{'$regex':cuisine_val ,'$options':'i'}}, 
                                                                      {'City':{'$regex':city_val ,'$options':'i'}}, {'Average Cost for two':{'$lte':int(price_val)}}]}}, {'$sort':{'Average Cost for two':1}}])
        
        calc_len = list(match_hotels)
        
        if (len(calc_len) == 0):
            cities = {'Restaurant Name':[0], 'Address':[0], 'Cuisines':[0], 'Average Cost for two':[0], 'Aggregate rating':[0]}
            cities_list = pd.DataFrame(cities, columns = ['Restaurant Name','Address','Cuisines','Average Cost for two','Aggregate rating'])
            
            
        else:
            hotel_dataframe = pd.DataFrame(calc_len)
            
            cities_list = hotel_dataframe[['Restaurant Name','Address','Cuisines','Average Cost for two','Aggregate rating']]
        
        
        count_of_hotels = len(cities_list.index)
        if count_of_hotels < 3:
            optional_cuisine = self.ninth_query(city_val)
            
            if optional_cuisine['Total Count'].max() > 3:
                new_data = optional_cuisine[optional_cuisine['Total Count'] == optional_cuisine['Total Count'].max()]
                cities_list = pd.concat([cities_list, new_data], ignore_index=True,sort=True)
                cities_list = cities_list.sort_index(ascending=False)
                cities_list = cities_list[['Restaurant Name','Address','Cuisines','Average Cost for two','Aggregate rating','Cuisine Type','Total Count']]
                cities_list = cities_list.fillna(0)
                cities_list.iloc[0, cities_list.columns.get_loc('Restaurant Name')] = 'Check for popular cuisine in this area'
                
        else:
            cities_list = cities_list.sort_index(ascending=False)
                
        
        return cities_list
    
    def ninth_query(self, get):
        '''Variety of hotels over the city'''
        list_of_hotels_count = []
        list_of_cuisine = self.third_query()
        list_of_cuisine = list_of_cuisine['Cuisines'].values.tolist()
        for cuisine in list_of_cuisine:
            list_of_hotels_count.append(self.col.aggregate([{'$match':{'$and':[{'Cuisines':{'$regex':cuisine, '$ne':None}},{'City':{'$regex':get,'$options':'i'}}]}},
                                                {'$group':{'_id':cuisine, 'count':{'$sum':1}}}, {'$sort':{'count':1}}]))
        
        counting_hotels = pd.DataFrame(list(list_of_hotels_count), columns=['List'])
        counting_hotels = counting_hotels['List'].apply(pd.Series)
        counting_hotels= counting_hotels.dropna().reset_index()
        counting_hotels.rename(columns={'_id':'Cuisine Type', 'count':'Total Count'}, inplace= True)
        counting_hotels = counting_hotels.sort_values(by=['Total Count'], ascending= False).reset_index()
        counting_hotels = counting_hotels.sort_index(ascending=False)
        counting_hotels = counting_hotels[['Cuisine Type','Total Count']]
        
        
        
        return counting_hotels
    
    
        
        
        
        
        
        
    
        
        
    
```
